```python
import pandas as pd
import os
import requests
```

```python
print(pd.__version__)
```

```python
token = ""

base_url = 'https://cloud.iexapis.com/v1'

resp = requests.get(base_url + '/status')

resp
```

```python
tickers= ["AAPL", "FB", "ADBE", "INTC", "PYPL", "BABA", "QCOM", "SHOP", "AMD", "MSFT"]
```

```python
sheet_requests = []

params = {'token': token,
          'period': "quarter",
         'last' : 24}

# requests.get(base_url+'/stock/AAPL/balance-sheet', params=params).json()

for ticker in tickers:
    
    endpoint = '/stock/{}/balance-sheet'.format(ticker)

    r = requests.get(base_url+endpoint, params=params)
    
    print(r)
    print(r.json())
    sheet_requests.append(r)
```

```python
sheet_data = pd.DataFrame(columns=list(sheet_requests[0].json()["balancesheet"][0].keys()))

for r in sheet_requests:
    for i in range(0, len(r.json()['balancesheet'])):
        mn = pd.DataFrame(r.json()['balancesheet'][i], index=[0])
        sheet_data = sheet_data.append(mn, ignore_index=True)

sheet_data.tail()
```

```python
sheet_data.to_csv("balance_sheet.csv", index=False)
```
